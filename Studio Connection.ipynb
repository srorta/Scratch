```python
import pandas as pd
import numpy as np
import time
from datetime import datetime
import cx_Oracle
import sqlalchemy
from sqlalchemy import create_engine
from pprint import pprint
```

```python
db = cx_Oracle.connect(dsn="P_S_SAM1") # creates database connection
cursor = db.cursor() #creates database cursor
query = """SELECT column_name FROM all_tab_cols WHERE table_name = 'DOVBOREHOLE' AND owner = 'GUYANA_INT' order by column_name""" #query that grabs all columns from DovBOREHOLE
res = cursor.execute(query) #Execute query
columns = [row[0] for row in res] #first row of columns from result

pprint(columns) #pretty print result
```

```python
#pandas does not like borehole properties because it is a blob. 
columns.remove('BOREHOLEPROPERTIES')
```

```python
sql_columns = ", ".join(col for col in columns) #creating variable with list of columns to return 
sql = f"Select {sql_columns} FROM GUYANA_INT.DOVBOREHOLE" #using f statement for sql query
```

```python
df = pd.read_sql_query(sql, con=db) # pandas has read_sql query function
```

```python
print(df.groupby(df['SPUDDATE'].dt.year))
```
