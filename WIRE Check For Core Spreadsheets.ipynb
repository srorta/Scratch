```python
#create requests session
session = requests.Session()
session.headers = headers
session.verify = False
```

```python
#Query for all wells with access group
group = 'guyana'
response = session.get(f"{base_url}/api/workflow/v2/wells/{group}/wells")
response.raise_for_status()
```

```python
print(response)
```

```python
#get well data
uwis = response.json()
type(uwis)
print(uwis)
```

```python
# Data Sections you want to retrieve
typesForRetrieval = 'surface_location'
```

```python
#for each well, get data
#for help with forming urls, use API
#surface location is the data type that has the well name. 
values = []
for uwi in uwis:
    #Get source data
    response = session.get('%s/api/workflow/v2/wells/%s?types=%s&display_units=true' % (base_url,uwi,typesForRetrieval))
    try:
        response.raise_for_status()
    except requests.HTTPError:
        print('Could not load data for %s' % uwi)
        continue
    resp = response.json()
    if resp is None:
        print('No surface location data for %s' % uwi)
    else:
        print(f"The surface location of {uwi} was added to the values list.")
        values.append(resp)
print(values)
```

```python
df2 = json_normalize(values, max_level = 2)
df2
```
