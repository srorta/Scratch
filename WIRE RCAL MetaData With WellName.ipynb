```python
#create requests session
session = requests.Session()
session.headers = headers
session.verify = False
```

```python
#Query for all wells with access group
group = 'guyana'
response = session.get(f"{base_url}/api/workflow/v2/wells/{group}/wells",verify=False)
response.raise_for_status()
```

```python
print(response)
```

```python
#get well data
uwis = response.json()
print(type(uwis))
print(uwis)
print(len(uwis))
```

```python
# Data Sections you want to retrieve
typesForRetrieval = 'rcal,surface_location,core_depth_shift,conventional_coring,sidewall_coring'
```

```python
#for each well, get data
#for help with forming urls, use API
values = []
for uwi in uwis:
    #Get source data
    response = session.get('%s/api/workflow/v2/wells/%s?types=%s&display_units=true' % (base_url,uwi,typesForRetrieval))
    try:
        response.raise_for_status()
    except requests.HTTPError:
        print('Could not load data for %s' % uwi)
        continue
    general = response.json()
    if 'rcal' not in general:
        print(f"No Rcal data found for {uwi}")
    if 'rcal' in general:
        print(f"Rcal data found for {uwi}")
        values.append(general)
    else:
        print(f"No Rcal data found for {uwi}")
        continue
print(f"{len(values)} wells with Rcal data found")
```

```python
df2 = json_normalize(values, max_level = 4)
df2
```
