```python
basefolder = r'C:\Users\srorta\Desktop\gapapps'
typesForRetrieval = 'final_general,final_depth,final_surface_location'
deviation = 'final_directional_survey'
deviation_list = []
values = []
none_uwi = []
none_deviation =[]
for uwi in uwis: 
    response3 = session.get('%s/api/workflow/v2/wells/%s?types=%s&display_units=true' % (base_url,uwi,deviation))
    response2 = session.get('%s/api/workflow/v2/wells/%s?types=%s&display_units=true' % (base_url,uwi,typesForRetrieval))
    response = session.get(f"{base_url}/api/workflow/v2/files/attachments?namespace=FormationEvaluationAndPressureData&uwi={uwi}")
    gapapps = response.json()
    file_pk_list = []
    print(uwi)
    for log in gapapps:
        if log['section_name'] == 'GAPAPPS':
            print(log['file_name'])
            file_pk = log['file_metadata_pk']
            file_pk_list.append(file_pk)
            max_pk = max(file_pk_list)
            print(file_pk_list)            
    for log in gapapps:
        if log['file_metadata_pk'] == max_pk:
            print(log['file_name'])
            print(log['download_url'])
            print(f"{base_url}{log['download_url']}")
            url = f"{base_url}{log['download_url']}"
            download_curve = session.get(url)
            with open(f"{basefolder}\\{log['original_file_name']}", "wb") as file:
                file.write(download_curve.content)
            try:
                response2.raise_for_status()
            except requests.HTTPError:
                print(f"Could not load data for {uwi}")
                continue
            response2Json = response2.json()
            if response2Json is None:
                values.append({})
                none_uwi.append(uwi)
            else:
                values.append(response2Json)
            
            response3Json = response3.json()
            if response3Json is None:
                deviation_list.append({})
                none_deviation.append(uwi)
            else:
                deviation_list.append(response3Json)
                

    file_pk_list.clear()
print(values)
print(none_uwi)
print(deviation_list)
```

```python
wire_metadata_df = json_normalize(values, max_level = 2)
wire_deviation_df = json_normalize(deviation_list)
wire_deviation_df
```
