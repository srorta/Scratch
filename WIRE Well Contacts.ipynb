```python
#create requests session
session = requests.Session()
session.headers = headers
session.verify = False
```

```python
#Query for all wells with access group
group = 'guyana'
response = session.get(f"{base_url}/api/workflow/v2/wells/{group}/wells")
response.raise_for_status()
```

```python
print(response)
```

```python
#get well data
uwis = response.json()
type(uwis)
print(uwis)
```

```python
# Data Sections you want to retrieve
typesForRetrieval = 'surface_location,contacts'
```

```python
#for each well, get data
#for help with forming urls, use API
values = []
for uwi in uwis:
    #Get source data
    response = session.get('%s/api/workflow/v2/wells/%s?types=%s&display_units=true' % (base_url,uwi,typesForRetrieval))
    try:
        response.raise_for_status()
    except requests.HTTPError:
        print('Could not load data for %s' % uwi)
        continue
    general = response.json()
    if general is None or 'contacts' not in general:
        print('No Contact Data for %s' % uwi)
    else: 
        print(f"Has Data {uwi}")
        values.append(general)
print(values)
```

```python
df2 = json_normalize(values)
df2
```

```python
df2.columns
```

```python
contacts_df = pd.DataFrame()
for x in df2.index.values:
    rcal_meta2 = pd.json_normalize(df2.iloc[x,0] ,errors = 'ignore')
    rcal_uwi2 = df2.iloc[x,7]
    well_name = df2.iloc[x,19]
    rcal_meta2["uwi"] = rcal_uwi2
    rcal_meta2["well_name"] = well_name
    contacts_df = contacts_df.append(rcal_meta2)
contacts_df
```
