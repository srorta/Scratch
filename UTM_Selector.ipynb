```python
import arcpy
import pandas as pd
arcpy.env.workspace = r"H:\UTM\CountryUTM.gdb"
arcpy.env.overwriteOutput = True
```

```python
fields = ["SHAPE@","COUNTRY_NA"]
```

```python
arcpy.MakeFeatureLayer_management(r"C:\Users\srorta\Desktop\Guyana\MyProject\MyProject.gdb\utm_merge_north_south","utm_layer")
```

```python
arcpy.MakeFeatureLayer_management(r"H:\UTM\utm.gdb\ihs_countries_ter_dissolve","country_layer")
```

```python
with arcpy.da.SearchCursor("country_layer",fields) as cursor:
    for row in cursor:
        print(row)
        geom = row[0]
        print(geom)
        arcpy.management.SelectLayerByLocation("utm_layer","INTERSECT",geom,'',"NEW_SELECTION",'')
        arcpy.management.CopyFeatures("utm_layer",f"H:\\UTM\\CountryUTM.gdb\\{row[1]}")
```

```python
fc_tables = arcpy.ListFeatureClasses()
for fc in fc_tables:
    print(str("processing" + fc))
    field = "FILENAME"
    expression = str(fc)
    arcpy.AddField_management(fc,field,"TEXT")
    arcpy.CalculateField_management(fc,field,'"'+expression+'"',"PYTHON")
```

```python
print(fc_tables)
```

```python
arcpy.management.Merge(fc_tables,"UTMCountries",'','')
```
