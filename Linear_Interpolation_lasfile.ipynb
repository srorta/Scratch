```python
las = lasio.read(r"C:\Users\srorta\Desktop\gapapps\bulletwood-1_gapapps.las")
```

```python
orig_las_df = las.df()
reset_index_df = orig_las_df.reset_index()
reset_index_df
```

```python
cds_df = pd.read_csv(r"C:\Users\srorta\Desktop\CDS\liza1bp2_postdrill-core_depth_shift.csv.csv")
cds_df
```

```python
cds_df["Delta(Core-Log) (m)"] = cds_df['Delta (Log - Core)  (m)'] *-1
cds_df
```

```python
coredepthlist = []
logdepthlist =[]
for i,row in reset_index_df.iterrows():
    #find high and low indices inside the CDS table
    logdepthlist.append(round(row['DEPTH'],2))
    searchindex = 0
    while searchindex < len(cds_df) and cds_df['Log Depth MD (m)'][searchindex] < row['DEPTH']:
        searchindex += 1
    lowindex = max(searchindex -1, 0)
    highindex = min(searchindex, len(cds_df)-1)
    print(lowindex)
    print(highindex)
    print(row['DEPTH'])
    
    #handle edge cases
    if lowindex == highindex:
        coredepth = None
        logdepth = None
        if (lowindex ==0):
            coredepth = cds_df['Core Depth MD (m)'][0]
            logdepth = cds_df['Log Depth MD (m)'][0]
        else:
            searchindex = len(cds_df)-1
            while cds_df['Core Depth MD (m)'][searchindex] is None and searchindex >0:
                searchindex -= 1
            coredepth = cds_df['Core Depth MD (m)'][searchindex]
            logdepth = cds_df['Log Depth MD (m)'][searchindex]
        delta = coredepth - logdepth
        coreshiftdepth = row['DEPTH'] + delta
        coreshiftdepth = round(coreshiftdepth, 2)
        coredepthlist.append(coreshiftdepth)
        print(coreshiftdepth)
    # handle interpolation 
    if lowindex != highindex:
        lowcoredepth = cds_df['Core Depth MD (m)'][lowindex]
        highcoredepth = cds_df['Core Depth MD (m)'][highindex]
        lowlogdepth = cds_df['Log Depth MD (m)'][lowindex]
        highlogdepth = cds_df['Log Depth MD (m)'][highindex]

        #if any of the above values are none, abort mission
        #if any(val is None for val in [lowCoreDepth, highCoreDepth, lowLogDepth, highLogDepth]):

        deltacoredepth = highcoredepth - lowcoredepth
        deltalogdepth = highlogdepth - lowlogdepth 
        factor = (row['DEPTH'] - lowlogdepth)/ deltalogdepth
        coreshiftdepth = factor * deltacoredepth + lowcoredepth
        coreshiftdepth =round(coreshiftdepth, 2)
        coredepthlist.append(coreshiftdepth)
        print(coreshiftdepth)
```

```python
depth_view_data = {"LogDepth": logdepthlist, "CoreDepth": coredepthlist}
depth_view_df = pd.DataFrame(depth_view_data)
depth_view_df
```

```python
depth_view_df["Delta(m)"] = depth_view_df["LogDepth"]-depth_view_df["CoreDepth"]
depth_view_df
```

```python
reset_index_df['DEPTH'] = depth_view_df['CoreDepth']
reset_index_df
```

```python
new_las_df = reset_index_df.set_index('DEPTH')
new_las_df
```

```python
las.set_data(new_las_df)
```
