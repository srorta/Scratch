```python
import pandas as pd
```

```python
main = pd.read_excel(r"H:\UTM\ISO-Country-UTM-Zones.xlsx")
ihs_countries = pd.read_csv(r"H:\UTM\ihs_countries_table.csv")
ihs_countries = ihs_countries.rename(columns={"COUNTRY_NAME":"ISO Country Name"})
ihs_countries["ISO Country Name"] = ihs_countries["ISO Country Name"].str.upper()
ihs_countries["ISO Country Name"] =ihs_countries["ISO Country Name"].str.replace(" ","")
main["ISO Country Name"] = main["ISO Country Name"].str.upper()
main["ISO Country Name"] = main["ISO Country Name"].str.replace(" ","")
```

```python
data = [main["ISO Country Name"]]
headers = ["ISO Country Name"]
df1 = pd.concat(data,axis=1,keys=headers)
df1["Type"] = "Excel"
data1 = [ihs_countries["ISO Country Name"]]
headers1 = ["ISO Country Name"]
df2 = pd.concat(data1,axis=1, keys=headers1)
df2["Type"] = "IHS"
df2
finaldf = pd.concat([df1,df2],ignore_index = True)
finaldf
```

The first .drop_duplicates() keeps the first instance of each unique row. The second drops any rows with the same value for ISO Country Name regardless of Type (keeping none of them). Because we are only dealing with unique rows at this point, the .drop_duplicates(subset = *, keep = False) necessarily drops only rows with duplicated col1 values and distinct col2 values. 

```python

finaldf.drop_duplicates().drop_duplicates(subset=["ISO Country Name"],keep=False)
```
