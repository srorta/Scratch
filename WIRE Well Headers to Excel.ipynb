```python
#get well data
uwis = response.json()
type(uwis)
print(uwis)
```

```python
# Data Sections you want to retrieve
typesForRetrieval = 'general,depth,surface_location'
```

```python
#for each well, get data
#for help with forming urls, use API
values = []
for uwi in uwis:
    #Get source data
    response = session.get('%s/api/workflow/v2/wells/%s?types=%s&display_units=true' % (base_url,uwi,typesForRetrieval))
    try:
        response.raise_for_status()
    except requests.HTTPError:
        print('Could not load data for %s' % uwi)
        continue
    general = response.json()
    if general is None or 'general' not in general:
        print('No well data for %s' % uwi)
        continue
    else:
        values.append(general)
print(values)
```

```python
df2 = json_normalize(values, max_level = 5)
df2
```
