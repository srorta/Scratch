```python
#Query to extract columns from borehole_,dovborehole, and position tables.
QUERY = '''SELECT 
                dbxobj.name NAME,
                dbxobj.folder_path FOLDER_PATH,
                dbxobj.display_type Display_Type,
                dbxobj.domain_object Domain_Object,
                t.x,
                t.y,
                t.z,
                t.w,
                t.id
            FROM GUYANA_INT.DBX_OBJECT dbxobj,
            table(sdo_util.getvertices(dbxobj.shape)) t
            
            WHERE GENERIC_DISPLAY_TYPE = 'Polygons' AND FOLDER_PATH LIKE '\Project\Input\Reference Datasets\Live Trace Outlines\%'
                
        '''
```

```python
#from sql query to pandas dataframe
df = pd.read_sql_query(sql = QUERY,con=connection)
df.columns = df.columns.str.upper()
df
```
